```python
import pandas as pd
from matplotlib import pyplot as plt
from wordcloud import WordCloud

%matplotlib inline
```

```python
df = pd.read_excel("data/FN_Training_Set.xlsx")
```

```python
df.Content = df.Content.astype(str)
```

```python
def plot_cloud(text):
    # Generate a word cloud image
    wordcloud = WordCloud().generate(text)

    # lower max_font_size
    wordcloud = WordCloud(max_font_size=40).generate(text)
    plt.figure(figsize=(14,8))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```python
plot_cloud(text = " ".join(df['Content Title'][df.click_bait_score==3]))
```

```python
plot_cloud(text = " ".join(df['Content Title'][df.click_bait_score==1]))
```

```python
plot_cloud(text = " ".join(df['Content'][df.click_bait_score==1]))
```

```python
plot_cloud(text = " ".join(df['Content'][df.click_bait_score==3]))
```
